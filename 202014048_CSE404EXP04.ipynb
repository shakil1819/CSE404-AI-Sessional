```python
class NQueens:
    def __init__(self, size):
        self.size = size
        self.solutions = 0
        self.solver()

    def solver(self):
        QPositions = [-1] * self.size
        self.put_queen(QPositions, 0)

        if self.solutions > 1:
            print(str(self.solutions) + " Solutions are possible for (" + str(self.size) + "x" + str(self.size) + ") grid.")
        elif self.solutions == 1:
            print(str(self.solutions) + " Solution is possible for (" + str(self.size) + "x" + str(self.size) + ") grid.")
        else:
            print("No solution is possible for (" + str(self.size) + "x" + str(self.size) + ") grid.")

    def put_queen(self, QPositions, target_row):
        if target_row == self.size:
            self.show_board(QPositions)
            self.solutions += 1
        else:
            for column in range(self.size):
                if self.check_constraints(QPositions, target_row, column):
                    QPositions[target_row] = column
                    self.put_queen(QPositions, target_row + 1)

    def check_constraints(self, QPos, row, col):
        for i in range(row):
            if QPos[i] == col or QPos[i] - i == col - row or QPos[i] + i == col + row:
                return False
        return True

    def show_board(self, QPositions):
        for r in range(self.size):
            row = ""
            for column in range(self.size):
                if QPositions[r] == column:
                    row += "Q"
                else:
                    row += "."
            print(row)
        print()

def main():
    N = int(input("Enter the number of test cases: "))
    for x in range(N):
        Q = int(input("Enter the grid size: "))
        if Q > 0:
            NQueens(Q)

if __name__ == "__main__":
    main()
```
